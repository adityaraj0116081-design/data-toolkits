2. Views instead of copies

. NumPy slicing usually returns a view on the original array (no data copied).

. Python list slicing always creates a new list (full copy).

```python
# 2. Write a Pandas script to find the mean of a specific column in a DataFrame.


import pandas as pd


data = {
    'name': ['A', 'B', 'C', 'D'],
    'marks': [70, 85, 90, 75]
}

df = pd.DataFrame(data)

# Calculate the mean of a specific column
mean_marks = df['marks'].mean()

print("Mean of marks column:", mean_marks)



```

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = {
    'maths': [70, 85, 90, 75],
    'science': [65, 80, 88, 78],
    'english': [72, 82, 91, 77]
}

df = pd.DataFrame(data)

corr_matrix = df.corr()
print(corr_matrix)

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# 7.  Write a program to perform element-wise multiplication of two NumPy arrays.

import numpy as np

# Create two NumPy arrays
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])

# Element-wise multiplication
result = arr1 * arr2

print("Array 1:", arr1)
print("Array 2:", arr2)
print("Element-wise multiplication:", result)


```

```python
# 8. Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [5, 6, 7, 8, 9]

# Plot multiple lines
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.plot(x, y3, label='Line 3')

# Add labels, title, and legend
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Line Plot with Multiple Lines")
plt.legend()

# Show plot
plt.show()
```

```python
# 9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

import pandas as pd

data = {
    'name': ['A', 'B', 'C', 'D'],
    'score': [65, 82, 90, 70]
}

df = pd.DataFrame(data)
print(df)

threshold = 75
filtered_df = df[df['score'] > threshold]

print(filtered_df)

```

```python
# 10.  Create a histogram using Seaborn to visualize a distribution.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'scores': [55, 60, 65, 70, 72, 75, 78, 80, 85, 90, 95]
}

df = pd.DataFrame(data)

# Create histogram
sns.histplot(df['scores'], bins=5, kde=True)

# Add labels and title
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.title("Distribution of Scores")

# Show plot
plt.show()
```

```python
# 11.  Perform matrix multiplication using NumPy.
import numpy as np

# Create two matrices
A = np.array([
    [1, 2],
    [3, 4]
])

B = np.array([
    [5, 6],
    [7, 8]
])

# Perform matrix multiplication
result = np.dot(A, B)
# or equivalently: result = A @ B

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Matrix multiplication result:\n", result)
```

```python
# 12. Create a 3D scatter plot using Plotly.


import plotly.express as px
import pandas as pd

# Sample data
data = {
    "x": [1, 2, 3, 4, 5],
    "y": [5, 4, 3, 2, 1],
    "z": [2, 3, 5, 7, 11]
}

df = pd.DataFrame(data)

# Create 3D scatter plot
fig = px.scatter_3d(
    df,
    x="x",
    y="y",
    z="z",
    title="3D Scatter Plot Example"
)

# Show plot
fig.show()

```
